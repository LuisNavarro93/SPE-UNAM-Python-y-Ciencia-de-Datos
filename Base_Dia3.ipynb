```python
def prod_stats_plt(data,log):
    fig, axes = plt.subplots(3,1, figsize=(8, 10), sharex=True)
    sns.histplot(data = data , x = log,ax=axes[0])
    axes[0].set_title("Histogram")

    sns.boxplot(data = data , x = log,ax=axes[1],showmeans=True)
    axes[1].set_xlabel("")
    axes[1].set_title("Boxplot")

    sns.violinplot(data = data , x = log,ax=axes[2],inner="quartile")
    axes[2].set_title("Violin plot")
    
    plt.tight_layout()
    plt.show()
```

```python
for param in lista_params:
    print("Boxplot para :",param)
    prod_stats_plt(df,param)
```
